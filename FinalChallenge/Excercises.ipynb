```python
### IMPORTANTE, CORRER PARA INSTALAR LAS LIBRERIAS NECESARIAS
import subprocess

def install_requirements():
    try:
        subprocess.check_call(['pip', 'install', '-r', 'requirements.txt'])
        print("Requerimientos instalados con exito.")
    except subprocess.CalledProcessError as ex:
        print(f"Error instalando los requerimientos: {ex}")
        raise

# Se llama la funcion para instalar los requerimientos
install_requirements()
```

```python
## SOLUCION

# Se importan las librerias necesarias
import math

# Definimos una clase para cada figura con sus respectivos metodos de calculo
class Triangle:
    def __init__(self, base:float, height:float):
        self.base = base
        self.height = height

    # Metodo para calcular el area
    def Area(self) -> float:
        try:
            # Hacemos el calculo de la formula base*h/2 con los parametros
            area = self.base * self.height / 2
            return area
        except TypeError as ex:
            # Retornamos error si algun parametro no tiene un tipo valido
            print(f"({ex}) ERROR: Parametro con tipo NO valido, favor ingresar un número")

class Rectangle:
    def __init__(self, width:float, height:float):
        self.width = width
        self.height = height

    # Metodo para calcular el area
    def Area(self) -> float:
        try:
            # Hacemos el calculo de la formula ancho*alto con los parametros
            area = self.width * self.height
            return area
        except TypeError:
            # Retornamos error si algun parametro no tiene un tipo valido
            print("ERROR: Debe ingresar números")

class Circle:
    def __init__(self, radius:float):
        self.radius = radius

    # Metodo para calcular el area
    def Area(self) -> float:
        try:
            # Hacemos el calculo de la formula pi*(r^2) con los parametros
            area = math.pi * math.pow(self.radius, 2)
            return area
        except TypeError:
            # Retornamos error si algun parametro no tiene un tipo valido
            print("ERROR: Debe ingresar números")

# Se instancian las clases de las diferentes figuras

def calculadoraGeometrica(opt:str):
    ## Triangulo
    if opt == '1':
        b = float(input("Ingrese la base del triangulo"))
        h = float(input("Ingrese la altura del triangulo"))
        triangulo = Triangle(b,h)
        triangleArea = triangulo.Area()
        if triangleArea:
            print(f"triangle's area is: {triangleArea:.1f}")
        return triangulo

    ## Rectangulo
    elif opt == '2':
        w = float(input("Ingrese el ancho del rectangulo"))
        h = float(input("Ingrese la altura del rectangulo"))
        rectangulo = Rectangle(w,h)
        rectangleArea = rectangulo.Area()
        if rectangleArea:
            print(f"rectangle's area is: {rectangleArea:.1f}")
        return rectangulo
    
    ## Circulo
    elif opt == '3':
        r = float(input("Ingrese el radio del circulo"))
        circulo = Circle(r)
        circleArea = circulo.Area()
        if circleArea:
            print(f"circle's area is: {circleArea:.1f}")
        return circulo
    
    else:
        print("ERROR: Opcion Invalida")
    
try:
    print("""
        Elije una figura:
        ╔═══════════════╗
        ║ 1. Triangulo  ║
        ║ 2. Rectangulo ║
        ║ 3. Circulo    ║
        ╚═══════════════╝ 
    """)
    calculadoraGeometrica(str(input("Ingrese una opción de la lista")))
except ValueError as err:
    print (f"({err}) ERROR: Parametro con tipo NO valido, favor ingresar un string")
```

2\) En una empresa, la cantidad de horas semanales trabajadas en el horario normal es de 36. La cantidad de horas trabajadas adicionales se consideran horas extras, y se pagan con un recargo del 25%.

Calcular el salario mensual obtenido por el trabajador conociendo los siguientes datos:

- Nombre
- Número de horas trabajadas en la semana
- Valor recibido por una hora normal de trabajo.

```python
## SOLUCION

# Calcular salario, en base a hrs semanales
def calcMonthlySalary(wagePerHour, hoursWorkedPerWeek):
    extraHours = int()
    extraSalary = int()
    wageperHourSurcharge = wagePerHour + (wagePerHour*0.25) # Salario por hora con recargo

    # Se valida si se hicieron horas extra
    if hoursWorkedPerWeek > 36:
        extraHours = hoursWorkedPerWeek-36  # -> Se calculan las horas extra
        extraSalary = extraHours * wageperHourSurcharge # -> Se calcula el pago de las horas extra
    
    regularWeeklySalary = hoursWorkedPerWeek * wagePerHour
    finalWeeklySalary = (hoursWorkedPerWeek * wagePerHour) + extraSalary # -> Se calcula el valor del pago semanal, más las horas extra
    monthlySalary = finalWeeklySalary * 4.34 # -> Se multiplica por las 4.34 semanas que tiene en promedio un mes

    # Se almacena la información a imprimir en un array/lista
    salaryInfo = [extraHours, hoursWorkedPerWeek, wagePerHour, wageperHourSurcharge, regularWeeklySalary, finalWeeklySalary, monthlySalary]

    return salaryInfo

try:
    name = str(input("Digita tu nombre"))
    pagoPorHora = int(input("Digita tu salario por hora"))
    horasTrabajadas = int(input("Digita las horas trabajadas por semana"))
            
    InfoSalario = calcMonthlySalary(pagoPorHora, horasTrabajadas)

    print(f"""
            {name}, esta es la info de tu salario:
            ╔══════════════════════════════╦═══════════════════»
            │  Horas semana laboral:       │  {36} 
            │  Horas extra realizadas:     │  {InfoSalario[0]:.1f}  
            │  Horas trabajadas en total:  │  {InfoSalario[1]:.1f} 
            ├──────────────────────────────┼───────────────────»
            │  Pago por hora:              │  ${InfoSalario[2]:.1f}  
            │  Pago por hora extra:        │  ${InfoSalario[3]:.1f} 
            │  Salario semanal base:       │  ${InfoSalario[4]:.1f}  
            │  Salario semanal total:      │  ${InfoSalario[5]:.1f} 
            │  Salario mensual:            │  ${InfoSalario[6]:.1f} 
            ╚══════════════════════════════════════════════════»
        """, end="")
except ValueError as err:
    print (f"ERROR: Parametro invalido, favor ingresar un string: ({err})")
```

3\)	Escriba un programa en Python que imprima los 89 primeros valores de la seria Fibonacci.

```python
## SOLUCION

# Definimos funcion que calcula n valores de la secuencia de fibonachi
def getFibonacciSecuence(n:int):
    # Se crea lista con n elementos
    secuencia:list =[1]*n
    indice = 2
    while (indice<n):
        # Se iteran los n elementos y se reemplazan por la suman de los valores de las dos secuencias anteriores 
        secuencia[indice] = secuencia[indice-2] +secuencia[indice-1]
        indice+=1
    return secuencia 

inpt = int()
while(not inpt):
    try:
        # Se pregunta al usuario cuantos valores de la secuencia desea ver, hasta que ingrese un valor valido
        inpt = int(input('Cuantos valores de la secuencia desea ver?: '))

        # Se valida que el número ingresado sea positivo
        if inpt > 0:
            # Se calcula la secuencia con la función y se imprime
            fiboSecuenceResult = getFibonacciSecuence(inpt) 
            print(f""" 
                Here you go, los primeros {inpt} valores de la Secuencia de Fibonacci: 
                {fiboSecuenceResult}
            """)
        else:
            # se retorna error si el número es negativo
            print(f"({inpt}) ERROR: Solo se permiten números positivos mayores a 0")

    except ValueError as ex:
        # Se retorna error si se ingresa un tipo de dato invalido
        print(f"ERROR: Parametro invalido, debe ingresar números enteros: ({ex})")
```
